`Question 1`. What is Flask Framework? What are the advantages of Flask Framework?

`Answer`:Flask is a popular Python web framework that allows developers to build web applications quickly and easily. It is a lightweight and flexible framework that provides developers with the essential tools needed to create web applications in a minimalistic manner.

Here are some advantages of Flask Framework:

1.`Lightweight`: Flask is a lightweight framework with no built-in layers of abstractions. It has a simple code base that is easy to understand, making it a popular choice among developers who want to create web applications quickly and efficiently.

2.`Flexibility`: Flask is an extremely flexible framework. It allows developers to choose the components they want to use, such as database integration, templating engine, or authentication, and integrate them easily into their web applications.

3.`Easy to Learn`: Flask has a small learning curve, making it easy for developers who are new to web development or Python to start building web applications. It also has excellent documentation, which makes it easy for developers to learn and get started with the framework.

4.`Minimalistic`: Flask is a minimalistic framework that does not include unnecessary features. This makes it easier to use and maintain, as there are fewer components to worry about.

5.`Extensible`: Flask is highly extensible, allowing developers to add new features and functionalities to their web applications easily. It also has a large community of developers who contribute to the development of various Flask extensions, making it easier for developers to add new features to their web applications.

4.`Testing`: Flask has an in-built test client that allows developers to test their web applications easily. This makes it easier to write and run tests, ensuring that the web application works as intended.

`Question 3`. What is App routing in Flask? Why do we use app routes?

`Answer `:In Flask, app routing is a way to define the endpoints or URLs that can be used to access different functionalities of your web application. Essentially, app routing allows you to map specific URLs to functions or views within your Flask application.

Using app routes in Flask has a number of benefits. First and foremost, it helps to organize your code and make it more modular. By defining routes, you can separate different parts of your application and handle them in distinct functions. This can make your code more maintainable and easier to update.

Additionally, app routes allow you to control the behavior of your application based on the user's input. By defining different routes for different functionality, you can ensure that your application responds appropriately to user requests. For example, you might define a route for a user to log in, another for a user to view their profile, and another for a user to submit a form.

Overall, app routing is a powerful feature of Flask that allows you to create dynamic and responsive web applications. By defining routes, you can organize your code, control the behavior of your application, and provide a smooth user experience.

`Question4`. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
`Answer:`
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    return '<h1>Company Name: ABC Corporation \n Location: India \n Contact Detail: 999-999-9999</h1>'


if __name__=="__main__":
    app.run(host="0.0.0.0")
```

`Question5` : What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
`Answer`:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the home page!'

@app.route('/user/<name>')
def user_profile(name):
    return f'Welcome to {name}\'s profile page!'

with app.test_request_context():
    # Generate a URL for the home page
    print(url_for('home'))  # Output: '/'

    # Generate a URL for a user profile page with name 'John'
    print(url_for('user_profile', name='John'))  # Output: '/user/John'
```
