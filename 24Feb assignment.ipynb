`Question 2`. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
DataFrame with a new index that starts from 1 and increments by 2 for each row.

`Answer`:

```python
df1 = pd.DataFrame({'A': list("RAHUL"),
                    'B': list("SURAJ"),
                    'C': list("PUSSB")})

def reindex_order(df):
    # get the number of rows in the DataFrame
    num_rows = df.shape[0]
    
    # create a new index starting from 1 and incrementing by 2 for each row
    new_index = [num for num in range(1, num_rows*2+1, 2)]
    
    return new_index
print(df1)
new_df = df1.reset_index(drop=True)
new_df.index = reindex_order(df1)
print("__________")
print(new_df)
```

`Question 5`. How are DataFrame.size() and DataFrame.shape() different?

`Answer`: Both `DataFrame.size()` and `DataFrame.shape()` are attributes of a Pandas DataFrame that provide information about the dimensions of the DataFrame. However, they are different in what they actually return.

`DataFrame.size()` returns the total number of elements in the DataFrame. For example, if a DataFrame has 3 rows and 4 columns, size() would return 12.

`DataFrame.shape()` returns a tuple of the dimensions of the DataFrame. The first element of the tuple is the number of rows in the DataFrame, and the second element is the number of columns. For example, if a DataFrame has 3 rows and 4 columns, shape() would return (3, 4).

In summary, `DataFrame.size()` returns the total number of elements in the DataFrame, while `DataFrame.shape()` returns a tuple of the dimensions of the DataFrame, specifically the number of rows and columns.

`Question 12`. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
import pandas as pd

# Create a sample DataFrame
df9 = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', end='2023-02-28', freq='D')
})

def select_rows_between_dates(df):
    start_date = '2023-01-01'
    end_date = '2023-01-31'
    
    mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    selected_rows = df.loc[mask]
    
    return selected_rows

selected_rows = select_rows_between_dates(df9)
selected_rows
```
